# DATA VISUALIZATION IN R USING GGPLOT2

**ggplot2** is a popular data visualization package in R based on the Grammar of Graphics. It provides a systematic way to create complex and customizable plots by layering components such as data, aesthetic mappings (x, y, color, size), geometric objects (points, bars, lines), and themes. Unlike base R graphics, ggplot2 is more flexible, produces publication-quality visuals, and makes it easier to add multiple layers, facets, and statistical transformations.

```r
install.packages("ggplot2")
library(ggplot2)
```

### Simple Bar Graph

```r
ggplot(mtcars, aes(x = factor(cyl))) +
  geom_bar(fill = "skyblue") +
  labs(title = "Bar Graph of Cylinders", x = "Cylinders", y = "Count")
```

### Stacked Bar Graph

```r
ggplot(mtcars, aes(x = factor(cyl), fill = factor(gear))) +
  geom_bar() +
  labs(title = "Stacked Bar Graph", x = "Cylinders", fill = "Gears")
```

### Grouped (Dodged) Bar Graph

```r
ggplot(mtcars, aes(x = factor(cyl), fill = factor(gear))) +
  geom_bar(position = "dodge") +
  labs(title = "Grouped Bar Graph", x = "Cylinders", fill = "Gears")
```

### Histogram

```r
ggplot(mtcars, aes(x = mpg)) +
  geom_histogram(binwidth = 3, fill = "steelblue", color = "black") +
  labs(title = "Histogram of MPG", x = "Miles per Gallon", y = "Count")
```

### Kernel Density Plot

```r
ggplot(mtcars, aes(x = mpg, fill = factor(cyl))) +
  geom_density(alpha = 0.5) +
  labs(title = "Kernel Density by Cylinders", x = "MPG")
```

### Box Plot

```r
ggplot(mtcars, aes(x = factor(cyl), y = mpg, fill = factor(cyl))) +
  geom_boxplot() +
  labs(title = "Box Plot of MPG by Cylinders",
       x = "Cylinders",
       y = "Miles per Gallon") +
  theme_minimal()
```

### Scatter Plot

```r
ggplot(mtcars, aes(x = hp, y = mpg, color = factor(cyl))) +
  geom_point(size = 3) +
  labs(title = "Scatter Plot of MPG vs Horsepower",
       x = "Horsepower",
       y = "Miles per Gallon",
       color = "Cylinders") +
  theme_light()
```

### Line graph

```r
# Create sample data
data <- data.frame(
  Month = factor(c("Jan", "Feb", "Mar", "Apr", "May", "Jun"),
                 levels = c("Jan", "Feb", "Mar", "Apr", "May", "Jun")),
  Sales = c(150, 200, 180, 220, 300, 250)
)

# Plot the line graph
ggplot(data, aes(x = Month, y = Sales, group = 1)) +
  geom_line(color = "blue", size = 1) +
  geom_point(color = "blue", size = 3) +
  labs(title = "Monthly Sales Over First Half of the Year",
       x = "Month", y = "Sales (in units)") 
```

## Summary

**ggplot2** is a powerful R package for creating high-quality graphics based on the Grammar of Graphics. It builds plots by combining layers:

**Data**  the dataset you want to visualize.

**Aesthetics (aes)**  how data variables are mapped to visual properties (x, y, color, size, shape).

**Geometries (geom_)**  the type of plot (bars, points, lines, boxes).

**Facets**  create multiple subplots by grouping variables.

**Themes**  control the appearance of the plot.
