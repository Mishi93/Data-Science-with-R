# Dimensionality Reduction

Dimensionality reduction is a technique used in data analysis and machine learning to reduce the number of variables (features) in a dataset while retaining as much important information as possible. It is especially useful when dealing with high-dimensional data to simplify models, reduce noise, and improve visualization.

Two common dimensionality reduction techniques are Principal Component Analysis (PCA) and Factor Analysis (FA).

### Principal Component Analysis (PCA)

**Concept:**

PCA is a method that transforms the original correlated variables into a smaller set of uncorrelated variables called principal components.

The first principal component captures the maximum variance in the data, the second captures the next highest variance orthogonal to the first, and so on.

PCA is widely used for feature reduction and visualization.



```r
# Sample dataset
data(iris)
df <- iris[, 1:4]  # Only numeric columns

# Perform PCA
pca_result <- prcomp(df, center = TRUE, scale. = TRUE)

# Summary of PCA
summary(pca_result)

# View principal components
pca_result$x

# Scree plot to visualize variance explained
plot(pca_result, type = "l")
```

### Factor Analysis (FA)

**Concept:**

Factor Analysis is a statistical method used to explain correlations between observed variables using a smaller number of latent factors.

Unlike PCA, which focuses on maximizing variance, FA assumes that observed variables are influenced by underlying unobserved factors plus error.

Commonly used in psychometrics, social sciences, and survey analysis.



```r
# Install package if not already installed
install.packages("psych")
library(psych)

# Perform Factor Analysis
fa_result <- fa(df, nfactors = 2, rotate = "varimax")  # 2 factors, varimax rotation

# View results
print(fa_result)

# Factor loadings
fa_result$loadings

# Plot factor diagram
fa.diagram(fa_result)
```

## Summary

| Feature               | PCA                                    | Factor Analysis (FA)                                        |
| --------------------- | -------------------------------------- | ----------------------------------------------------------- |
| Goal                  | Reduce dimensionality, retain variance | Identify latent factors explaining correlations             |
| Components vs Factors | Principal Components                   | Latent Factors                                              |
| Assumption            | No underlying model                    | Observed variables are influenced by latent factors + error |
| Focus                 | Maximize variance                      | Explain covariance structure                                |
